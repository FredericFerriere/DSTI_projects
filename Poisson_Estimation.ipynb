```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# Parameter Estimation for a Poisson Distribution

**Methodology:**<br>
*We're generating numSimul samples each of size sampleSize.*<br>
*For each simulation, the parameter is estimated using the method of moments (order 1)*

```python
poissonParam = 5
numSimul = 1000
sampleSize = 200

numpy.random.seed(1234)
fullSample = np.random.poisson(lam=poissonParam, size=(numSimul,sampleSize))
estimations = fullSample.mean(1)
```

#### Checking the distribution of values we get from our simulations

```python
sampleDensity = plt.hist(estimations, bins=13, density=True)
plt.title("Distribution of estimated parameters")
plt.xlabel("parameter")
plt.ylabel("density")
plt.show()
```

#### The distribution looks gaussian, we'll compare it to the "equivalent" gaussian (with mean and standard deviations calculated from sample estimations

```python
estimationsMean = estimations.mean()
estimationsStd = estimations.std()
normalDensity = norm.pdf(sampleDensity[1],estimationsMean,estimationsStd)
```

```python
sampleDensity = plt.hist(estimations, bins=13, density=True, color='blue', alpha=0.5, label="Sample Density")
plt.title("Distribution of estimated parameters")
plt.xlabel("parameter")
plt.ylabel("density")
plt.plot(sampleDensity[1], normalDensity, label="Normal Density", color='red')
plt.legend(loc='upper right')
plt.show()
```
